## Funcion desanidar columnas 

```python
def str_to_list(value):
     ## F que convierte una cadena de texto en una lista 
    if pd.isna(value): 
        ## recibe como parametro un valor que es la cadena de texto a convertir 
        ## La función utiliza la función pd.isna(value) de Pandas para verificar si el valor es NaN (no un número). 
        return [] 
        # Si es así, significa que la cadena de texto es nula, y en ese caso, la función retorna una lista vacía [].
    else:
        value = ast.literal_eval(value) ## si el valor no es NaN, a función utiliza la función ast.literal_eval(value) para evaluar la cadena de texto y convertirla en un objeto de Python
        if isinstance(value, bool) or isinstance(value, float):
            return [] ## Por ultimo verifica si el obj, es de tipo bool o float, si es asi retorna una lista vacia 
        return value ## de lo contrario asume que no lo es y por lo tanto es una lista valida  
```

```python
def str_to_dict(value):
    ## F que convierte una cadena de texto en un dicc
    if pd.isna(value):
        return None
    else:
        return ast.literal_eval(value)
```

## Funcion para obtener elementos de una lista de lista de dicc

```python
def get_list(funct, dataframe='', column='', key_name=''):
    ## funct se utiliza convertir una cadena en una lista o diccionario, 
    ## dataframe, que es el DataFrame del cual se extraerá la información, 
    ## column, que es el nombre de la columna del DataFrame que contiene los datos relevantes,
    ## key_name, que es el nombre de la clave del diccionario cuyos valores se desean obtener.
    
    dataframe[column] = dataframe[column].apply(funct)
    column_list = dataframe[column].to_list()

    return_list = [[data[key_name] for data in inter_list if key_name in data] for inter_list in column_list]

    return return_list
```

```python
df_movies['belongs_to_collection'] = df_movies['belongs_to_collection'].apply(str_to_dict)
```

```python
df_desanidado = pd.json_normalize(df_movies['belongs_to_collection']).fillna('')
df_desanidado.head()
```

```python
columns = df_desanidado.columns
new_columns_names = {column : f'belongs_to_collection_{column}' for column in columns} ## renombro a las columnas 
df_desanidado.rename(columns=new_columns_names, inplace=True)
df_desanidado.head()
```

```python
#Elimino las columnas que no voy a utilizar, para luego poder hacer un merge con el dataF principal y solo quedarme con la columna name
df_desanidado.drop(columns=['belongs_to_collection_id', 'belongs_to_collection_poster_path', 'belongs_to_collection_backdrop_path'], inplace=True)
```

```python
df_movies = pd.concat([df_movies, df_desanidado], axis=1) ## hago un merge con el Df princial
df_movies.drop(columns='belongs_to_collection', inplace=True) ## luego elimino la columna que ya no me sirve
```

```python
columns = ['genres', 'production_companies', 'production_countries', 'spoken_languages']
key = 'name'
df = pd.DataFrame()
for _, column in enumerate(columns):
    elements_list = []
    elements_list = get_list(str_to_list, dataframe=df_movies, column=column, key_name=key)
    ## llamo a la funcion que hice arriba para convertir un srt en una lista 
    df_aux = pd.DataFrame({column: elements_list})
    df = pd.concat([df, df_aux], axis=1)
```

```python
df_movies.drop(columns=columns, inplace=True)
```

```python
df_movies = pd.concat([df_movies, df], axis=1)
df_movies.head()
```

## ETL dataset Credits

```python
df_credits = pd.read_csv('Dataset/credits.csv')
```

```python
columns = ['cast', 'crew']
key_list = ['name', 'name']
df_actors = pd.DataFrame()
for _, (column, key) in enumerate(zip(columns, key_list)):
    elements_list = []
    elements_list = get_list(str_to_dict, dataframe=df_credits, column=column, key_name=key)
    if 'crew' in column:
        new_list = [element[0] if len(element) > 0 else 'No Director' for element in elements_list]
        elements_list = new_list.copy()
    df_aux = pd.DataFrame({column: elements_list})
    df_actors = pd.concat([df_actors, df_aux], axis=1)
df_actors = pd.concat([df_actors, df_credits['id']], axis=1)
df_actors.head()
```

```python
df_movies_new = pd.merge(df_movies_nuevo, df_actors_nuevo, how='left', on='id')
```
